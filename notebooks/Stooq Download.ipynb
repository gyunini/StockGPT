```python
import pandas as pd
```

```python
from bs4 import BeautifulSoup as bs4
```

```python
import random
import re
import requests
import time
import tqdm
```

```python
us_url = "https://stooq.com/t/?i=518&v=0&l={page_num}"
```

```python
us_universe = []
```

```python
for page_num in tqdm.tqdm(range(1, 85)):
    page_url = us_url.format(page_num=page_num)
    response = requests.get(page_url)
    response.raise_for_status()
    page = response.text

    stock_name_pattern = re.compile(r"^q/\?s=.*$")

    soup = bs4(page, 'html.parser')
    name_links = soup.find_all('a', href=stock_name_pattern)
    stock_names = [name_link.get_text(strip=True) for name_link in name_links]
    valid_names = [name for name in stock_names if name.split('.')[-1] == 'US']

    if page_num != 84:
        assert len(valid_names) == 100

    us_universe += valid_names

    time.sleep(random.uniform(0,2))
```

```python
us_universe
```

```python
download_url = "https://stooq.com/q/d/l/?s={stock_name}&i=d"

for stock_name in (pbar:=tqdm.tqdm(us_universe)):
    pbar.set_description(f"Downloading {stock_name}")
    url = download_url.format(stock_name=stock_name)
    response = requests.get(url)
    response.raise_for_status()

    with open(f"stooq/{stock_name}.csv", 'wb') as fp:
        fp.write(response.content)

    time.sleep(random.uniform(0,2))
```
